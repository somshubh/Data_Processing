```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np

df=pd.read_csv('data.csv')
#df['Country'].isnull().sum()
df
```

# Removing the null value rows

```python
df.dropna()
```

# Using simpleImputer numerical value 

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
imputer.fit(df.iloc[:,1:3].values)
df.iloc[:,1:3]=imputer.transform(df.iloc[:,1:3].values)
df
```

```python
df['Country'].value_counts
```

# catagorical null value handling

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:]=imputer.transform(df.iloc[:,:].values)
df
```

# Null value handling for google play store value

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df=pd.read_csv('googleplaystore.csv')
df
```

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3]=imputer.transform(df.iloc[:,2:3].values)
df
```
