# Relative Frequency

```python
Country_car=pd.crosstab(df['origin'],'Numbers of Car')
Country_car['relative_frequency']=Country_car['Numbers of Car']/Country_car['Numbers of Car'].sum()
Country_car
```

# Persentage Frequency

```python
Country_car['persentage_frequency']=Country_car['relative_frequency']*100
Country_car
```
