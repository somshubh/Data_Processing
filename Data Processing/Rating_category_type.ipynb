```python
import pandas as pd
import numpy as np

df=pd.read_csv('googleplaystore.csv')
df
```

### How many free Apps present in ART_AND_DESIGN CATEGORY

```python
df=df.values
```

```python
c=0
for i in df:
    if(i[1]=='ART_AND_DESIGN' and i[6]=='Free'):
      c+=1
print(c,"apps are ART_AND_DESIGN Category and free type")
```

### How many Apps in ART_AND_DESIGN CATEGORY with more than 4.5 rating

```python
c=0
for i in df:
    if(i[1]=='ART_AND_DESIGN' and i[2]>4.5):
      c+=1
print(c,"apps are ART_AND_DESIGN Category with more than 4.5 Rating")
```

#### How many  Apps are there in FAMILY CATEGORY with more than 4.5 rating and Free

```python
c=0
for i in df:
    if(i[1]=='FAMILY' and i[2]>4.5 and i[6]=='Free'):
      c+=1
print(c,"apps are FAMILY Category with more than 4.5 Rating and free type")
```
