```python
import pandas as pd
import numpy as np
import seaborn as sns

df=pd.read_csv('/kaggle/input/heart-desease-dataset/Heart-disease/heart.csv')
df.head()
```

## Distribution Plot for numerical columns

```python
sns.displot(df['Age'],kde=True)
```

```python
sns.distplot(df['Age'],kde=True,hist=True,rug=True,bins=25)
```

```python
sns.distplot(df['RestingBP'],kde=True,color='Green')
```

```python
sns.distplot(df['Cholesterol'],kde=True,color='brown')
```

```python
sns.distplot(df['MaxHR'],kde=True,color='blue')
```

## Categorical columns pie chart visulization
